```python
profit_path = '/Users/ryan/Documents/GitHub/MGT4187-Project/data'

profit = pd.read_csv('movie_profit.csv')
```

```python
movie_lst = profit['movie'].tolist()
filtered_data_name = basics[basics['primaryTitle'].isin(movie_lst)]
```

Merge data

```python
# merge with profit data
combined_data = pd.merge(filtered_data_name, profit, left_on='primaryTitle', right_on='movie', how='left')
```

```python
combined_data['release_date'] = pd.to_datetime(combined_data['release_date'], format='%m/%d/%Y')

# filter entries where the year is greater than 2000
filtered_df_release = combined_data[combined_data['release_date'].dt.year > 2000]
```

```python
# merge with ratings data
filtered_df_release = pd.merge(filtered_df_release, ratings, on='tconst', how='left')

# sort the data by title and number of votes
df_sorted = filtered_df_release.sort_values(by=['primaryTitle', 'numVotes'], ascending=[True, False])

# drop duplicates, keeping the first entry (highest 'numVotes') for each 'primaryTitle'
deduped_df = df_sorted.drop_duplicates(subset='primaryTitle', keep='first')
```
